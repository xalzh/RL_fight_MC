```python
import pandas as pd
import matplotlib.pyplot as plt

# Read the CSV file
csv_files = ['results/results_1zombie.csv', 'results/results_2zombie.csv', 'results/results_3zombie.csv']

for csv_file in csv_files:
    data = pd.read_csv(csv_file, header=None)
    epochs = list(data.iloc[:, 0])
    rewards = list(data.iloc[:, 1])
    plt.plot(epochs, rewards)
    plt.xlabel('Epochs')
    plt.ylabel('Rewards')
    plt.title('Rewards vs Epochs')
    plt.grid()
    plt.show()
```
